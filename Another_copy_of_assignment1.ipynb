```python
import numpy as np
import pandas as pd
#list
list1 = [1, 2, 3, 4, 5]
list2 = [6, 7, 8, 9, 10]
#2. Arrays
array1 = np.array(list1)
array2 = np.array(list2)
#3. Identify their type using type()
print("Type of list1:", type(list1))
print("Type of array1:", type(array1))
#4. Perform mathematical operations on these datasets created - multiplication, division,power of
print("Multiplication:", array1 * array2)
print("Division:", array1 / array2)
print("Power off:", np.power(array1, array2))
print(f"Addition of Two:\narray1 + array2\n{array1 + array2}")
#6. Implement np.sin(), log(), log2(), np.exp())
print("Sin:", np.sin(array1))
print("Log:", np.log(array1))
print("Log2:", np.log2(array1))
print("Exp:", np.exp(array1))
```

```python
import matplotlib.pyplot as plt
people =['kiran', 'arun', 'vijay', 'varun']
age =[25, 30, 35, 40]
height =[145, 151, 165, 173]
weight=[45, 55, 65, 75]
plt.scatter(weight,height)
plt.title('relationship w and h')
plt.ylabel('h i mtrs')
plt.xlabel('w in kgs')
plt.show()
# Scatter Plot
plt.scatter(age, height, label='Height')
plt.scatter(age, weight, label='Weight')
plt.title('Scatter Plot')
plt.xlabel('Age')
plt.ylabel('Value')
plt.legend()
plt.show()
# Bar Chart
plt.bar(people, height, label='Height')
plt.bar(people, weight, label='Weight', bottom=height)
plt.title('Bar Chart')
plt.xlabel('People')
plt.ylabel('Value')
plt.legend()
plt.show()
# Histogram
plt.hist(age, bins=5, edgecolor='black')
plt.title('Histogram')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```
